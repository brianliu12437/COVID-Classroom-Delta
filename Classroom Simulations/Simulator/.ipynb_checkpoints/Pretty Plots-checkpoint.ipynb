```python
import pandas as pd
import sys
import warnings
import numpy as np
import matplotlib.pyplot as plt
import pickle
if not sys.warnoptions:
    warnings.simplefilter("ignore")
import seaborn as sns
import gc
import itertools
import random
sys.path.append('../')
```

```python
nomask_distanced = pd.read_csv('updated_function_nomasks_distanced.csv')
nomask_distanced.drop('Unnamed: 0',axis = 1,inplace = True)
nomask_distanced['percent_cap'] = round(nomask_distanced["N"]/294,2)
cond = (nomask_distanced['percent_cap']==0.49) & (nomask_distanced['Percent Vaccinated'] == 0.5)
nomask_distanced_toplot = nomask_distanced[cond].groupby('feet_apart').infected.mean().reset_index()

mask_distanced = pd.read_csv('updated_function_masks_distanced.csv')
mask_distanced.drop('Unnamed: 0',axis = 1,inplace = True)
mask_distanced['percent_cap'] = round(mask_distanced["N"]/294,2)
cond = (mask_distanced['percent_cap']==0.49) & (mask_distanced['Percent Vaccinated'] == 0.5)
mask_distanced_toplot = mask_distanced[cond].groupby('feet_apart').infected.mean().reset_index()


mask_clumped = pd.read_csv('clumpy_masked.csv')
mask_clumped.drop('Unnamed: 0',axis = 1,inplace = True)
mask_clumped['percent_cap'] = round(mask_clumped["N"]/294,2)

no_mask_clumped = pd.read_csv('clumpy_no_masked.csv')
no_mask_clumped.drop('Unnamed: 0',axis = 1,inplace = True)
no_mask_clumped['percent_cap'] = round(no_mask_clumped["N"]/294,2)
no_mask_clumped['Percent Vaccinated'] = no_mask_clumped['p']

mask_random =  pd.read_csv('random_mask_final.csv')
mask_random.drop('Unnamed: 0',axis = 1,inplace = True)
mask_random['percent_cap'] = round(mask_random["N"]/294,2)

no_mask_random = pd.read_csv('random_nomask_final.csv')
no_mask_random.drop('Unnamed: 0',axis = 1,inplace = True)
no_mask_random['percent_cap'] = round(no_mask_random["N"]/294,2)
```

```python
mask_clumped['percent_cap'].unique()
```

```python
percent_cap = 0.98

mask_clumped_toplot = mask_clumped[mask_clumped['percent_cap']== percent_cap]
no_mask_clumped_toplot = no_mask_clumped[no_mask_clumped['percent_cap']== percent_cap]

mask_random_toplot =mask_random[mask_random['percent_cap']== percent_cap]
no_mask_random_toplot =no_mask_random[no_mask_random['percent_cap']== percent_cap]

mask_clumped_toplot = mask_clumped_toplot.groupby('Percent Vaccinated').infected.mean().reset_index()
no_mask_clumped_toplot =  no_mask_clumped_toplot.groupby('Percent Vaccinated').infected.mean().reset_index()

mask_random_toplot = mask_random_toplot.groupby('Percent Vaccinated').infected.mean().reset_index()
no_mask_random_toplot = no_mask_random_toplot.groupby('Percent Vaccinated').infected.mean().reset_index()


mask_distanced_toplot = mask_distanced[mask_distanced['percent_cap'] == percent_cap]
mask_distanced_toplot = mask_distanced_toplot.groupby(['Percent Vaccinated','feet_apart']).infected.mean().reset_index()
mask_distanced_toplot = mask_distanced_toplot.groupby('Percent Vaccinated').infected.min().reset_index()

nomask_distanced_toplot = nomask_distanced[nomask_distanced['percent_cap'] == percent_cap]
nomask_distanced_toplot = nomask_distanced_toplot.groupby(['Percent Vaccinated','feet_apart']).infected.mean().reset_index()
nomask_distanced_toplot = nomask_distanced_toplot.groupby('Percent Vaccinated').infected.min().reset_index()

```

```python
plt.figure(figsize = (12,10))



fig, ax = plt.subplots(figsize = (12,10))

plt.plot(no_mask_clumped_toplot['Percent Vaccinated'],no_mask_clumped_toplot['infected'],label = 'clumped no mask',color = 'purple')
plt.plot(mask_clumped_toplot['Percent Vaccinated'],mask_clumped_toplot['infected'],label = 'clumped mask', color = 'purple', linestyle = 'dashed')

plt.plot(no_mask_random_toplot['Percent Vaccinated'],no_mask_random_toplot['infected'],label = 'random no mask', color = 'blue')
plt.plot(mask_random_toplot['Percent Vaccinated'],mask_random_toplot['infected'],label = 'random mask', color = 'blue',linestyle = 'dashed')


plt.plot(nomask_distanced_toplot['Percent Vaccinated'],nomask_distanced_toplot['infected'],label = 'maximum distancing no mask', color = 'green')
plt.plot(mask_distanced_toplot['Percent Vaccinated'],mask_distanced_toplot['infected'],label = 'maximum distancing mask', color = 'green',linestyle = 'dashed')

labels = ['0%','0%','20%','40%','60%','80%','100%']
ax.set_xticklabels(labels)
plt.legend()
plt.xlabel('Percent of Population Vaccinated', labelpad = 10)
plt.ylabel('Average Number of Secondary Infections', labelpad = 10)
plt.title('100% Classroom Occupancy')
```
